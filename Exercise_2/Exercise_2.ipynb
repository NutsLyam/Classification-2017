**10.** Создать матрицу:

$$
\begin{pmatrix}
0 & 1 & 2 & 3 & 4 \\
1 & 2 & 3 & 4 & 0 \\
2 & 3 & 4 & 0 & 1 \\
3 & 4 & 0 & 1 & 2 \\
4 & 0 & 1 & 2 & 3
\end{pmatrix}
$$

```python
def foo(a):
    x = np.arange(a)
    for i in range(1,a):
        addition = np.hstack((np.arange(i,a),np.arange(0,i)))
        x = np.vstack((x, addition))
    return x

print (foo(6))   
```
